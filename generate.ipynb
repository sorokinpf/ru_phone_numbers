```python
import re
import pandas as pd
```

```python
data = open('numbers.txt').read()
```

```python
print (data)
```

```python
groups = re.findall('<tr>.*?</tr>',data,re.DOTALL)[1:]
```

```python
s = []
for g in groups:
    s.append(re.findall('<td.*?>([^<]*)</td>',g))
    
```

```python
phones = pd.DataFrame(s,columns=['name','code','start','end','date'])
```

```python
phones
```

```python
phones['start'] = phones.start.astype(int)
phones['end'] = phones.end.astype(int)
```

```python
phones['pull'] = phones.end - phones.start + 1
```

```python
phones.pull.sum()
```

```python
# example : "phone":"+7123) 123-12-32"
format_phone = '+7{code}) {}{}{}-{}{}-{}{}\n'
def format_number(code,number):
    number = '%07d'%number
    return (format_phone.format(*number,code=code))
```

```python
done = 0
for i,row in phones.iterrows():
    print ('starting ',i,' done: ',done)
    f = open('numbers/%d(%s)_numbers.txt'%(i,row.code),'w')
    for number in range(row.start,row.end+1,1):
        if number % 100000 == 0:
            print (row.code,number)
        f.write(format_number(row.code,number))
    done += row.end+1 - row.start
    f.close()
```
